```python
seq1 = Seq('ATGCATGGTGCGCGA')
seq2 = Seq('ATTTGTGCTCCTGGA')
```

```python
#Global Alignment
alignments = pairwise2.align.globalxx(seq1,seq2)
```

```python
#view all
for a in alignments:
  print(format_alignment(*a))
```

```python
#Local Alignment
loc_alignments = pairwise2.align.localxx(seq1,seq2)
```

```python
# view all
for a in loc_alignments:
  print(format_alignment(*a))
```

```python
#hamming Distance fxn
def hamming_distance(lhs,rhs):
  return len([(x,y) for x,y in zip(lhs,rhs) if x!=y])
```

```python
def delta(x, y):
    return 0 if x == y else 1

def M(seq1, seq2, i, j, k):
    return sum(delta(x, y) for x, y in zip(seq1[i:i+k], seq2[j:j+k]))

def makeMatrix(seq1, seq2, k):
    n = len(seq1)
    m = len(seq2)
    return [[M(seq1, seq2, i, j, k) for j in range(m-k+1)] for i in range(n-k+1)]

def plotMatrix(M, Psi, seq1, seq2, nonblank='', blank=' '):
    print(' ' + seq2 * 2)
    print(' ' * (2 + len(seq2)) + '|' + ' ' * (len(seq2) - 1))
    for label, row in zip(seq1, M):
        line = ''.join(nonblank if s < Psi else blank for s in row)
        print(label + '|' + line)

def dotplot(seq1, seq2, k=1, t=1):
    M = makeMatrix(seq1, seq2, k)
    plotMatrix(M, t, seq1, seq2)

# Contoh pemanggilan fungsi dotplot
seq1 = "ATCGATCG"
seq2 = "ATGGTCGA"
dotplot(seq1, seq2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Add Some fancyness to it
plt.imshow(np.array(makeMatrix(seq1, seq2, 1)))
xt = plt.xticks(np.arange(len(list(seq1))), list(seq1))
yt = plt.yticks(np.arange(len(list(seq2))), list(seq2))
plt.show()
```

```python
# Convert to Fxn
def dotplotx(seq1, seq2):
    plt.imshow(np.array(makeMatrix(seq1, seq2, 1)))
    # on x-axis list all sequences of seq 2
    xt = plt.xticks(np.arange(len(list(seq2))), list(seq2))
    # on y-axis list all sequences of seq 1
    yt = plt.yticks(np.arange(len(list(seq1))), list(seq1))
    plt.show()
```

```python
dotplotx(seq1,seq2)
```
